```python
import requests
from selenium import webdriver
from BeautifulSoup import BeautifulSoup
```

```python
## Javascript inside html, therefore requests alone is not able to load this dynamic page
url = 'http://finance.yahoo.com/quote/NTES/options?p=NTES'
driver = webdriver.Chrome()   # Chrome web driver from selenium
driver.get(url)  # load from that url
page = driver.page_source  # load the page content from the web driver
```

```python
soup = BeautifulSoup(page)
```

```python
print soup.prettify()
```
